## Estatística e remoção de dados 

```python
features = df.columns.drop("Class").to_list()
coluna_target = "Class"
analise_mw = {}
classes = df[coluna_target].unique()
alfa = 0.01
agrupamento = df.groupby(coluna_target)
```

```python
# verificando qual features tem um valor de significansia menos que 0.01
for coluna in features:
    grupo0 = agrupamento.get_group(classes[0])[coluna]
    grupo1 = agrupamento.get_group(classes[1])[coluna]
    
    estatistica_mw, pvalue_mw = mannwhitneyu (grupo0,grupo1)
    analise_mw[coluna]={
        "pvalue" : pvalue_mw,
        "estatistica" : estatistica_mw,
        "relação_segnificativa":pvalue_mw < alfa, 
    } 
```

```python
df_mw = pd.DataFrame(analise_mw).T

df_mw
```

```python
df_mw.query("relação_segnificativa == False")
```

## Discução sobre a redução de dados da nossa base 




### Limpeza e Preparação dos Dados

Iremos retirar as features que possuem significância menor que 1% da nossa base, com o objetivo de melhorar o treinamento do modelo.

Além disso, removeremos alguns outliers da base de dados. Como nosso conjunto é bastante desbalanceado — menos de 1% das transações correspondem a fraudes —, os outliers serão removidos apenas da classe sem fraudes.

Essa abordagem busca evitar a distorção dos dados e proporcionar uma melhor interpretação e desempenho do modelo.

```python
# separando os DataFrame em classe 1 e 0
df_majoritaria = df[df["Class"]==0]
df_minoritaria = df[df["Class"]==1]

df_majoritaria.shape, df_minoritaria.shape
```

```python
# criando uma função para remoção outliers 
def remover_outliers_quantil(df, variaveis, quantil_inferior=0.05, quantil_superior=0.95):
    df_filtrado = df.copy()

    for variavel in variaveis:
        limite_inferior = df[variavel].quantile(quantil_inferior)
        limite_superior = df[variavel].quantile(quantil_superior)
        df_filtrado = df[(df[variavel]>= limite_inferior) & (df[variavel]<= limite_superior)]
    return df_filtrado
```

```python
variaveis = df.columns.drop("Class")
```

```python
# removendo os outliers da classe maioritária
df_majoritaria_filtrado = remover_outliers_quantil(df_majoritaria, variaveis)
df_majoritaria_filtrado.shape
```

```python
# embaralhando a base de dados e resetando o index
df_sem_outliers = pd.concat([df_majoritaria, df_minoritaria])
df_sem_outliers = df_sem_outliers.sample(frac=1, random_state=42).reset_index(drop=True)
df_sem_outliers
```

```python
# dropando as features com uma significansia menos que 0.01
df_sem_outliers = df_sem_outliers.drop(
    columns = df_mw.query("relação_segnificativa == False").index
) 
```

# Discuções

Ao analisar os boxplots após a remoção dos outliers, observa-se que as escalas dos dados foram reduzidas, indicando uma melhor distribuição das variáveis. Além disso, é importante destacar que a proporção entre as classes não se alterou de forma significativa, mantendo menos de 1% das observações pertencentes à classe 1, correspondente às transações em que houve ocorrência de fraude.
